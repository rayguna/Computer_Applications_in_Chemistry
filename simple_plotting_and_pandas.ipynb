# Simple Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'x': [i for i in range(11)]})

def f_quadratic(x):
  return x**2

df['y'] = df['x'].apply(f_quadratic)

df.head() #check data
```

# Simple Plotting

```python
df.plot(x='x', y='y', kind='line')
plt.show()
```
